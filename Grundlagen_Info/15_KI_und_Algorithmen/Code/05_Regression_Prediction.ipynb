```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error

df = pd.read_csv('../Data/apartment_prices.csv')
X = df[['Wohnflaeche_m2', 'Zimmer', 'DistanzZentrum_km', 'Baujahr']]
y = df['Preis_CHF']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred) ** 0.5
print('MAE:', round(mae, 1))
print('RMSE:', round(rmse, 1))
```

```python
plt.figure(figsize=(6,4))
plt.scatter(y_test, y_pred)
plt.xlabel('Echte Preise')
plt.ylabel('Vorhergesagte Preise')
plt.title('Prediction: echt vs. vorhergesagt')
plt.show()
```
